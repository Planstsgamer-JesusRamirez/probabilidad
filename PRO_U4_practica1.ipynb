```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Definimos la media y la desviación estándar
media = 70
desviacion_estandar = 4

# Generamos datos aleatorios de la distribución normal
datos = stats.norm.rvs(loc=media, scale=desviacion_estandar, size=500)

# Calculamos la probabilidad de estudiantes que pesan más de 68 y menos de 72 kg
z_inf = (68 - media) / desviacion_estandar
z_sup = (72 - media) / desviacion_estandar

probabilidad = stats.norm.cdf(z_sup) - stats.norm.cdf(z_inf)

# Obtenemos el total de estudiantes
total_estudiantes = 500

# Calculamos la cantidad de estudiantes que pesan más de 68 y menos de 72 kg
cantidad_estudiantes = probabilidad * total_estudiantes

# Graficamos la distribución normal
x = np.linspace(media - 5 * desviacion_estandar, media + 5 * desviacion_estandar, 200)
y = stats.norm.pdf(x, loc=media, scale=desviacion_estandar)

plt.axvline(x=68, color='red', linestyle='dashed', linewidth=1, label='68 kg')
plt.axvline(x=72, color='red', linestyle='dashed', linewidth=1, label='72 kg')
plt.plot(x, y, label='Distribución normal')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.title('Distribución de pesos de estudiantes')
plt.legend()
plt.show()

# Imprimimos la cantidad de estudiantes
print("Cantidad de estudiantes:", int(cantidad_estudiantes))
```
